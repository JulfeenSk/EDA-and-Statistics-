## 2.Take sample from population 

```python
#Different size of sample
sample_size = [10,30,100] # different sample size
sample_size
```

```python
# Sample mean and confidence Interval

sample_mean = [] # empty list
confidence_interval = []

for i , sample_size in enumerate(sample_size):
    sample_mean_i = []
    for _ in range(1500):
        #no of samples - the more the sample it becomes normal dist eventually
        sample = np.random.choice(population_score, size=sample_size)
        sample_mean_i.append(np.mean(sample))
    sample_mean.append(sample_mean_i)   
```

```python
# Visual the Sampling distribution

plt.hist(sample_mean_i , bins= 20, color ='orange')
plt.xlabel('Sample mean')
plt.ylabel("Frequency")
plt.title(f'sampling Distribution (Sample Size{sample_size:})')
```

```python
# Estimate the populationmean 

sample_mean = np.mean(sample_mean_i)
sample_std = np.std(sample_mean_i,ddof = 1)
standard_error = sample_std / np.sqrt(sample_size)

print('Sample Mean :', sample_mean)
print('Sample Std:', sample_std)
print('Sample Error :', standard_error)
```

### Confidence Interval

```python
z_critical = norm.ppf(0.975)
z_critical
```

```python
margin_of_error = z_critical * standard_error
margin_of_error
```

```python
confidence_interval=(sample_mean - margin_of_error, sample_mean + margin_of_error)
confidence_interval
```

**Problem**  : Customer Satisfaction Survey for a Retail Store
    
Imagine you are tasked with conducting a customer satisfcation survey for a retail store with hundreds of locations . The goal is ti estimate the overrall satisfaction level of customers with the stores services and products . However , visiting and surveying every customer at every store is time_consuming  and expensive.     Therefore, you need to employ statistical techniques to efficiently collect and analyze data

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd
```

### Generate data

```python
np.random.seed(42)
data = pd.DataFrame({
    'Store_ID' : np.arange(1,1001),
    'Cust_Satisfaction' : np.random.randint(1,11,size=1000)
})
```

```python
data
```

### Simple Random Sampling 

```python
sample_size = 50 
random_sample = data.sample(n=sample_size, random_state=42)
```

```python
random_sample
```

```python
random_sample.head()
```

### Stratified Sampling

```python
strata = data.groupby('Store_ID').size().reset_index()
strata.columns=['Store_ID','Count']
strata['Stratum']= pd.cut(strata['Count'],bins=[0,200,400,600,800,1000], labels=['1','2','3','4','5'])
stratified_sample = data.groupby('Store_ID',group_keys=False).apply(lambda x : x.sample(min(len(x),2),random_state=42))
```

```python
strata.columns
```

```python
strata
```

```python
stratified_sample
```

### Cluster Sampling

```python
cluster_size = 5 
clusters = data.groupby(np.arange(len(data))//cluster_size)
cluster_sample = clusters.apply(lambda x : x.sample(1,random_state=42))
```

```python
len(clusters)
```

```python
cluster_sample
```

### Visualize the sampling methods 

```python
sns.set(style='whitegrid')
fig,axes = plt.subplots(3,1,figsize=(15,15))

sns.histplot(data,x='Cust_Satisfaction', ax=axes[0], kde = True)
axes[0].set_title('Full Data Distribution')

sns.histplot(random_sample,x='Cust_Satisfaction', ax=axes[1], kde = True)
axes[1].set_title('Simple Random Sample')

sns.histplot(stratified_sample,x='Cust_Satisfaction', ax=axes[2], kde = True)
axes[2].set_title('Stratified Sample')

plt.show()
```

```python
# For cluster Sampling 
cluster_sample_plot = px.hostpgram(cluster_sample, x ='Cust_Satisfaction',title='Cluster Sampling')
cluster_sample_plot.(update_traces(marker = dict(color = 'rgba(0,123,255,0.6)'))
cluster_sample_plot.update_layout(bargap = 0.1)
cluster_sample_plot.show()                     
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generate_population(population_size):
    return np.random.normal(loc= 50,scale= 10,size = population_size )
# the population will be in the normal form
```

```python
def random_sampling(population_size , sample_size,num_samples):
    return [np.random.choice(population,size=sample_size,replace=False) for _ in range(num_samples)]
```

```python
#for _ in range(100):
    #print(_)
# creates an anonymus function
```

```python
def analyze_visualize(population,samples):
    
    population_mean = np.mean(population)
    sample_means = [np.mean(sample) for sample in samples]
    
    plt.figure(figsize=(12,8))
    
    plt.subplot(2,2,1)
    plt.hist(population,bins=30,color ='red',alpha = 0.7)
    plt.title("Population Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    
    plt.subplot(2,2,2)
    plt.hist(sample_means,bins=30,color ='green',alpha = 0.7)
    plt.title("Sampling Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    
    #Comparision
    plt.subplot(2,2,4)
    plt.hist(sample_means,bins=30,color ='green',alpha = 0.7)
    plt.axvline(population_mean, color ='red',linestyle='dashed',label='Population_mean') # line graph
    plt.title("Sampling Mean adn Population Mean Distribution")
    plt.xlabel("Sample MEan")
    plt.ylabel("Frequency")
    
    plt.tight_layout()
    plt.show()
```

```python
def  main():
    population_size  = 3000
    sample_size = 30
    num_samples = 30
    
    population = generate_population(population_size)
    samples= random_sampling(population , sample_size , num_samples)
    analyze_visualize(population,samples)
```

```python
# main excecutes 1st 
if __name__ == "__main__":
    main()
```

### Title : The effect pf a new teaching method on student performance 
**Objective:** `To determine wheather a new teaching method leads to improved academic performance comparedd to traditional teaching method in a specific subject area.`

**Hypothesis:** `The new teaching method significantlyimproves student performance compared to the traditional teaching method`

## 1.Data Collection

```python
import pandas as pd
import random
```

```python
# Generating sample adta for the traditioal teaching method
random.seed(0)
traditional_scores = [random.randint(50,100) for _ in range(100)]

# Generating sample adta for the traditioal teaching method
random.seed(1)
new_teaching_scores = [random.randint(60,100) for _ in range(100)]

#Creating a dAtframe to store the data 
data = pd.DataFrame({
    'Traditional_Teaching_Scroes': traditional_scores,
    'New_Teaching_Scores' : new_teaching_scores  
})

#Saving the data to a csv file
data.to_csv('student_performance_data.csv', index=False)
```

```python
df=pd.read_csv('student_performance_data.csv')
```

```python
df
```

## 2.Sampple Selection

```python
sample_size = 25
sample_data = df.sample(sample_size)
```

```python
sample_data
```

## 3.Formulation of Hypothesis

**H0** New teaching method has not improved students performance significantly

**H1** New teaching method has improved students performance significantl

## 4.Hypothesis Testing

```python
Traditional_Scores= sample_data['Traditional_Teaching_Scroes']
New_scores = sample_data['New_Teaching_Scores']
```

```python
print(Traditional_Scores.mean())
```

```python
New_scores.mean()
```

```python
# Perform some tests
alpha = 0.05
```

```python
from scipy import stats

#Performing the t-test
t_statistic, p_value = stats.ttest_ind(traditional_scores,new_teaching_scores)
```

```python
import numpy as np 
print("Mean of Traditioanl_Teaching_Scores = ",np.mean(sample_data['Traditional_Teaching_Scroes']))
print("Mean of New_Teaching_Scores = ",np.mean(sample_data['New_Teaching_Scores']))
```

```python
# Compare the p-value with significance level


if p_value < alpha:
    print("Reject the null Hypothesis. Accept  the Alternate Hypothesis")
else:
    print("Accept the Null Hypothesiis")
```
