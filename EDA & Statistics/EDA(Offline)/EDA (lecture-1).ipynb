```python
import requests 
#Library used to fetch details from a website
import bs4

fetchdetails = requests.get("https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012")

```

```python
container =bs4.BeautifulSoup(fetchdetails.text,"lxml")
```

```python
cases=container.find_all("div",class_="lister-item mode-advanced")
```

```python
print(len(cases))
```

```python
year=[]
Runtime=[]
Moviename=[]
for i in cases :
    movieyear=i.find("span",class_="lister-item-year")
    year.append(movieyear.string)
    duration=i.find("span",class_="runtime")
    Runtime.append(duration.string)
    moviename=i.find("img")["alt"]
    Moviename.append(moviename)
print(year) 

```

```python
print(Runtime)
```

```python
print(Moviename)
```

```python
import pandas as pd
imdb_dataset = pd.DataFrame({"Movie":Moviename ,"Duration":Runtime,"MovieReleaseyear":year})
imdb_dataset
```
