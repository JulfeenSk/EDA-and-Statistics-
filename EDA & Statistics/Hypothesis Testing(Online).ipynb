Suppose you are a  data analyst for an e-commerce company and you are interested in testing whether a recent website redesugn has significantly increased the average time spent by usera on the website . YOu collect data on the time spwnt by a  random sample of users before and after the redesign . Now lets perform hypothesis tesing using Python

```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from scipy import stats 
```

```python
# generate syntethic data 

np.random.seed(42)
before_redesign = np.random.normal(5,2,100) #(mean,std,count)
after_redesign = np.random.normal(6,2,100)
```

```python
sns.histplot(before_redesign,kde=True)
```

```python
sns.histplot(after_redesign,kde=True)
```

```python
#visualization made easy
sns.set(style='whitegrid')
sns.displot(before_redesign,kde=True)
sns.displot(after_redesign,kde=True)
plt.show()
```

```python
# Perform a t test 
t_stat , p_value = stats.ttest_ind(before_redesign,after_redesign)
alpha = 0.05 # significance level
```

```python
print("T-stats : ",t_stat)
print("p_value : ",p_value)
```

```python
if p_value <alpha:
    print("Reject the null Hypothesis.YEs, there is more time spending after redesign")
else:
    print("Accept the null Hypothesis.No,there is no more time spending after redesign")
```
