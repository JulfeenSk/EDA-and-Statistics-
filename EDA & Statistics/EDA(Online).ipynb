### Bivariate Analysis

#### 1.Numerical to Numerical

```python
#Scatter plot

plt.scatter(data['ApplicantIncome'],data['LoanAmount'])
plt.xlabel('Applicant Income')
plt.ylabel('LoanAmount')
plt.show()
```

```python
sns.scatterplot(data=data,x='ApplicantIncome', y ='LoanAmount')
```

```python
#Joint Plot using Seaborn 

sns.jointplot(data=data,x='LoanAmount', y='ApplicantIncome', kind='scatter')
plt.suptitle('Joint Plot')
```
