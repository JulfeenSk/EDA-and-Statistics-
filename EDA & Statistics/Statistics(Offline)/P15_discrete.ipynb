```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
import numpy as np
```

```python
# Poisson Diistribution

from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3, size=10000)
sns.histplot(data_poisson, bins=30, kde=False, color='skyblue', element='step', fill=True)
plt.xlabel('Poisson Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Binomial Distribution

from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
sns.histplot(data_binom, kde=False, color='orange', element='step', fill=True)
plt.xlabel('Binomial Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bernoulli Distribution

from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6)
sns.histplot(data_bern, kde=False, color='green', element='step', fill=True)
plt.xlabel('Bernoulli Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Geometric Distribution

from scipy.stats import geom
data_geom = geom.rvs(p=0.2, size=10000)  # p is the probability of success
sns.histplot(data_geom, kde=False, color='blue', element='step', fill=True)
plt.xlabel('Geometric Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Hypergeometric Distribution

from scipy.stats import hypergeom
N = 100  # Total number of items
K = 30   # Number of successful items
n = 10   # Number of draws
data_hypergeom = hypergeom.rvs(M=N, n=K, N=n, size=10000)

sns.histplot(data_hypergeom, kde=False, color='purple', element='step', fill=True)
plt.xlabel('Hypergeometric Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Negative binomial distribution

from scipy.stats import nbinom
n = 5  # Number of successes required
p = 0.2  # Probability of success in each trial
data_negative_binomial = nbinom.rvs(n, p, size=1000)
sns.histplot(data_negative_binomial, bins=30, kde=False, color='brown', element='step', fill=True)
plt.xlabel('Negative Binomial Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Negative Hypergeometric Distribution

from scipy.stats import nhypergeom
M = 20  # Total number of items in the population
N = 10  # Number of successful items in the population
n = 5   # Number of trials
x = np.arange(0, n+1)
data_nhyper = nhypergeom.pmf(x, M, N, n)
plt.bar(x, data_nhyper, align='center', alpha=0.7)
plt.xlabel('Number of Successes in ' + str(n) + ' Trials')
plt.ylabel('Probability')
plt.title('Negative Hypergeometric Distribution PMF')
plt.grid(True)
plt.show()
```
