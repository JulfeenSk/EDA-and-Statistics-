## BIVARIATE ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bank.csv")
df.head()

```

```python
# NUMERICAL - NUMERICAL ASSOCIATIONS
plt.scatter(df.balance,df.duration)
plt.show()
```

```python
# Scatter Plot
df.plot.scatter(x="age",y="balance")
plt.show()
```

```python
# Line Plot

x = df.balance
y = df.duration
plt.plot(x, y)
plt.show()
```

```python
# Correlation Matrix
sns.heatmap(df[["age","balance"]].corr(), annot=True, cmap="Blues")
```

```python
# CATEGORICAL - NUMERICAL ASSOCIATIONS
# Box Plot
sns.boxplot(data=df, x="balance", y="housing")
plt.show()
```

```python
# Bar Graph
df.groupby("loan")["balance"].mean().plot.bar()
```

```python

# Count Plot
sns.countplot(data = df, x = 'marital', hue = 'loan')
```

```python
# CATEGORICAL - CATEGORICAL ASSOCIATIONS
# Crosstabs
# used to represent the frequency of categorical values - college
cross_tab_prop = pd.crosstab(index = df['marital'], columns = df['loan'], normalize = "index")
cross_tab_prop
```

```python
# Stacked Bar Graphs
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

plt.legend(loc="upper right", ncol=2)
plt.xlabel("Marital")
plt.ylabel("Loan") 
plt.show()
```

```python
# Stacked Bars with Labels
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab10', figsize=(10, 6))

plt.legend(loc="upper right", ncol=2)
plt.xlabel("Marital")
plt.ylabel("Loan")

for n, x in enumerate([*cross_tab_prop.index.values]):
    for l in cross_tab_prop.loc[x]:
                
        plt.text(x=n,
                 y=l,
                 s=f'{np.round(l * 100, 1)}%', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")

plt.show()
```
