```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
```

```python
# discrete
# Bernoulli Distribution
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000,p=0.6) #p =success rate i.e 0=40% ,1 =60%
sns.histplot(data_bern, kde=False, color='green', element='step', fill=True)
plt.xlabel('Bernoulli Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
#discrete
# Uniform Distribution


from scipy.stats import uniform
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
sns.histplot(data_uniform, bins=100, kde=True, color='skyblue', element='step', fill=True)
plt.xlabel('Uniform Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
#discrete
# Poisson Distribution

from scipy.stats import poisson
data_poisson = poisson.rvs(mu=3, size=10000)
sns.histplot(data_poisson, bins=10, kde=False, color='skyblue', element='step', fill=True)
plt.xlabel('Poisson Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
#discrete
# Binomial Distribution

from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,size=10000)
sns.histplot(data_binom, kde=False, color='orange', element='step', fill=True)
plt.xlabel('Binomial Distribution')
plt.ylabel('Frequency')
plt.show()
```
