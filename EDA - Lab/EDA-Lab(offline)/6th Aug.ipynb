# Bivariate  Analysis

```python
# To change the Column names
mobile_data2 = mobile_data.rename(columns = {
    'Rating ?/5':'Ratings',
    'Price in INR' : 'Price',
    'Date of Scraping' : 'Date',
    'ROM/Storage' : 'ROM',
})
mobile_data2.info()
```

```python
# Replace the values
mobile_data2["RAM"].replace({
    "2GB": "2 GB RAM",
    "4GB": "4 GB RAM",
    "6GB": "6 GB RAM",
    "8GB": "8 GB RAM",
    "12GB": "12 GB RAM"
    
}, inplace=True)
mobile_data2
```

```python
mobile_data2.isnull().sum()
```

```python
# Removing all rows which are having missing values.
mobile_data2.dropna()

# Removing only that rows where entire columns have missing values.
mobile_data2.dropna(how="all")

# Remove columns with any null values
mobile_data_cleaned = mobile_data2.dropna(axis=1)
```

```python
mobile_data2.isnull().sum()
```

```python
#filling missing values with backward fil
df1 = mobile_data2.fillna(method='bfill')
print(df1)

# filling missing values using forward fill
df2 =  mobile_data2.fillna(method='ffill')
print(df2)

#filling missing values using mode
mobile_data2.ROM = mobile_data2.ROM.fillna(mobile_data2.ROM.mode()[0])

#filling missing values using mode
mobile_data2.Ratings = mobile_data2.Ratings.fillna(mobile_data2.Ratings.mean())
```

```python
mobile_data2.isnull().sum()
```

```python
sns.kdeplot(x='Ratings', hue='ROM', data=mobile_data2).set(title="Relationship between the Rom and ratings",)
plt.show()
```

```python
sns.barplot(x='Ratings', y='ROM', data=mobile_data2)
```

```python
mobile_heatmap = mobile_data2.corr()

sns.set(rc={'figure.figsize':(8,5)})
sns.heatmap(mobile_heatmap, annot=True, cmap='rocket')
plt.show()

#
mobile_corr=mobile_data2.corr()
mobile_corr
```
