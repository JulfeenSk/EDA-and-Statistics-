```python
fig, axes = plt.subplots(2,2)

plt.figure(figsize=(6, 4)) 

axes[0,0].boxplot(mobile['Ratings'])
axes[0,0].title('Box Plot of Ratings')

axes[0,1].boxplot(mobile['Number of Ratings'])
axes[0,1].title('Box Plot of Number of Ratings')

axes[1,0].boxplot(mobile['RAM'])
axes[1,0].title('Box Plot of RAM')

axes[1,1].boxplot(mobile['Ratings'])
axes[1,1].title('Box Plot of ROM')

plt.show()
```

```python
mobile['RAM'] = mobile['RAM'].str.extract(r'(\d+)').astype(float)
mobile['ROM'] = mobile['ROM'].str.extract(r'(\d+)').astype(float)
mobile['Ratings']= mobile['Ratings'].astype(float)
```
